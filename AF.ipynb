```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

# Activation Functions

```python
def logistic_sigmoid(x):
    return 1/(1+math.exp(-x))
def d_logistic_sigmoid(x):
    return logistic_sigmoid(x)*(1-logistic_sigmoid(x))

def scaled_sigmoid(x):
    return 4 * logistic_sigmoid(x) - 2
def d_scaled_sigmoid(x):
    return 4 * d_logistic_sigmoid(x)

def improved_sigmoid(x, alpha=7, a=0.01):
    if x >= a:
        return (alpha * (x-a)) + logistic_sigmoid(a)
    elif x <= -a:
        return (alpha * (x+a)) + logistic_sigmoid(-a)
    else:
        return logistic_sigmoid(x)
def d_improved_sigmoid(x, alpha=7, a=0.01):
    if abs(x) >= a:
        return alpha
    else:
        return d_logistic_sigmoid(x)

def SiLU(x):
    return x * logistic_sigmoid(x)
def d_SiLU(x):
    return logistic_sigmoid(x) * (1 + x * (1 - logistic_sigmoid(x)))

def soft_root_sign(x, alpha=1, beta=1):
    return x / (x/alpha + math.exp(-x/beta))
def d_soft_root_sign(x, alpha=1, beta=1):
    return (1 + x/beta) * math.exp(-x/beta) / (x/alpha + math.exp(-x/beta)) ** 2

def double_sigmoid(x, sigma=6):
    if x < 0:
        return logistic_sigmoid(x + sigma)
    else:
        return logistic_sigmoid(-x + sigma)
def d_double_sigmoid(x, sigma=6):
    if x < 0:
        return d_logistic_sigmoid(x + sigma)
    else:
        return - d_logistic_sigmoid(-x + sigma)
```

```python
x = np.linspace(-20,20,1000)
```

```python
plt.figure(figsize=(5,3))
plt.plot(x, [double_sigmoid(v) for v in x], c='black', label='AF(x)')
plt.plot(x, [d_double_sigmoid(v) for v in x], c='black', label="AF'(x)", linestyle='--')
plt.xlim((-20,20))
plt.ylim((-0.3,1.05))
plt.xlabel('x')
plt.ylabel('f')
plt.legend()
plt.grid(True)
plt.tight_layout()
```

```python
plt.figure(figsize=(5,3))
plt.plot(x, [logistic_sigmoid(v) for v in x], c='black', label='AF(x)')
plt.plot(x, [d_logistic_sigmoid(v) for v in x], c='black', label="AF'(x)", linestyle='--')
plt.xlim((-20,20))
plt.ylim((-0.3,1.05))
plt.xlabel('x')
plt.ylabel('f')
plt.legend()
plt.grid(True)
plt.tight_layout()
```

```python
plt.figure(figsize=(5,3))
plt.plot(x, [scaled_sigmoid(v) for v in x], c='black', label='AF(x)')
plt.plot(x, [d_scaled_sigmoid(v) for v in x], c='black', label="AF'(x)", linestyle='--')
plt.xlim((-20,20))
plt.ylim((-2.5,2.5))
plt.xlabel('x')
plt.ylabel('f')
plt.legend()
plt.grid(True)
plt.tight_layout()
```

```python
plt.figure(figsize=(5,3))
plt.plot(x, [improved_sigmoid(v, alpha=0.01, a=7) for v in x], c='black', label='AF(x)')
plt.plot(x, [d_improved_sigmoid(v, alpha=0.01, a=7) for v in x], c='black', label="AF'(x)", linestyle='--')
plt.xlim((-20,20))
plt.ylim((-0.2,1.2))
plt.xlabel('x')
plt.ylabel('f')
plt.legend()
plt.grid(True)
plt.tight_layout()
```

```python
plt.figure(figsize=(5,3))
plt.plot(x, [SiLU(v) for v in x], c='black', label='AF(x)')
plt.plot(x, [d_SiLU(v) for v in x], c='black', label="AF'(x)", linestyle='--')
plt.xlim((-10,10))
plt.ylim((-0.5,10))
plt.xlabel('x')
plt.ylabel('f')
plt.legend()
plt.grid(True)
plt.tight_layout()
```

```python
plt.figure(figsize=(5,3))
plt.plot(x, [soft_root_sign(v, alpha=1, beta=0.4) for v in x], c='black', label='AF(x)')
plt.plot(x, [d_soft_root_sign(v) for v in x], c='black', label="AF'(x)", linestyle='--')
plt.xlim((-20,20))
plt.ylim((-0.3,1.05))
plt.xlabel('x')
plt.ylabel('f')
plt.legend()
plt.grid(True)
plt.tight_layout()
```
